```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
P0 = 1
Q = 1
R = 1000

# Lists to store values for plotting
Pt_list = [P0]
Kt_list = [0]
Ptplus1_t_list = [P0]

# Iterate Kalman equations
for _ in range(1, 1000):
    Pt_tminus1 = Ptplus1_t_list[-1]
    
    Kt = Pt_tminus1 / (Pt_tminus1 + R)
    Pt = (1 - Kt) * Pt_tminus1
    Ptplus1_t = Pt + Q

    Kt_list.append(Kt)
    Pt_list.append(Pt)
    Ptplus1_t_list.append(Ptplus1_t)

# Plot a posteriori and a priori error covariances
plt.figure(figsize=(10, 5))
plt.plot(Pt_list, label='a posteriori: $P_t$')
plt.plot(Ptplus1_t_list, '--', label='a priori: $P_{t+1|t}$')
plt.xlabel('Time step')
plt.ylabel('Error Covariance')
plt.legend()
plt.title('Kalman Filter Error Covariances')
plt.grid(True)
plt.show()

# Plot Kalman gain
plt.figure(figsize=(10, 5))
plt.plot(Kt_list, label='Kalman Gain: $K_t$')
plt.xlabel('Time step')
plt.ylabel('Gain')
plt.legend()
plt.title('Kalman Gain vs. Time')
plt.grid(True)
plt.show()
```
