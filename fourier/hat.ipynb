```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.001
L = np.pi
x = L * np.arange(-1, 1+dx, dx)
n = len(x)
nquart = int(np.floor(n/4))
```

```python
f = np.zeros_like(x)
f[nquart:2*nquart] = 1 + x[nquart:2*nquart]/(L/2)
f[2*nquart:3*nquart] = 1 - x[2*nquart:3*nquart]/(L/2)
plt.plot(x, f)
```

```python
Ff = np.zeros_like(x)
N = 100
a = np.zeros(N)
Ff_arr = np.zeros((N, len(x)))

a_0 = np.sum(f*dx)
for k in range(N):
    a[k] = np.sum(f * np.cos(np.pi*(k+1)*x/L))*dx

Ff += a_0/2
for k in range(N):
    Fk = a[k]*np.cos((np.pi*(k+1))*x/L)
    Ff += Fk
    Ff_arr[k] = Ff
    plt.plot(x, Fk, '--')

plt.plot(x, Ff, 'r')
```
