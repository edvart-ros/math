```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.001
t = np.arange(0, 1, dt)
n = len(t)

f1 = 50
f2 = 120
f = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)
plt.plot(t, f)
```

```python
noise = 2.5*np.random.randn(n)
g = f + noise
plt.plot(t, g)
plt.plot(t, f)
```

```python
f_hat = np.fft.fft(f, n)
plt.plot(f_hat)
```

```python
PSD = f_hat*np.conj(f_hat)/n
freq = (1/(dt*n))*np.arange(n)
L = np.arange(1, np.floor(n/2), dtype='int')
indices = PSD > 100
PSDclean = PSD * indices
f_hat = indices * f_hat
ffilt = np.fft.ifft(f_hat)
plt.plot(t, ffilt)
print("denoised data")
```
