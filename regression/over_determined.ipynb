```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_fn(x, A, b, lambda1):
    return (np.linalg.norm(A@x-b) + lambda1*np.linalg.norm(x, ord=1))
```

```python
n, m = 500, 100
A = np.random.rand(n, m)
b = np.random.rand(n)

xdag = np.linalg.pinv(A)@b
lambda1s = np.array([0, 0.1, 0.5])
```

```python
fig, axs = plt.subplots(len(lambda1s), 2)

for j in range(len(lambda1s)):
    res = minimize(cost_fn, args=(A,b,lambda1s[j]), x0=xdag)
    x = res.x
    axs[j,0].bar(range(m),x)
    axs[j,0].set_ylabel('lam='+str(lambda1s[j]))
    axs[j,1].hist(x,20)
    axs[j,1].set_xlim(-0.15,0.15)
```
